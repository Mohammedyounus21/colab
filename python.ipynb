```python
class Person:
  #initializer/constructor for objects
  def __init__(self, name, age): # self is represented as object to be created
    self.name = name   # Object Variable #instance attributes
    self.age = age     # Object Variable

# Creating Objects
per1 = Person("vishwa", 99)  # Object 1
per2 = Person("shivani", 97)  # Object 2

# Accessing Object Variables
print(per1.name)  # vishwa
print(per2.name)  # shivani
```

```python
#static method
class person:
  country="india"

  @classmethod #decorator-meta data / additional inforamation which goes and tell python the function which is defined has additional capabilities
  def greet(cls):
    print("hello from the",cls.country) # have acess to the class property and no access to the object property

  @staticmethod # used to keep it isolated from other attributes  # is independent - no acees to any class or object properties
  def hello():
    print("hello")

  def __init__(self,name,age):
    self.name=name
    self.age=age

  def findage(self):
    return self.age

per=person("vishwa",99)
print(per.findage())
person.greet()
per.greet() #can be called through class and object both
person.hello()
```

```python
class parent:
  property=90 # direct values

  def eat(self):
    print("parent eating")

class child(parent):
  property=89

  def display(self):
    print("child property",self.property)
    print("parent property",super().property)

  def eat(self):
    print("child eating")

  def calleat(self):
    self.eat()  #child eat
    super().eat() # parent's eat

obj=child()
obj.display()
obj.calleat()
```

```python
class num:
  def __init__(self,num):
    self.num=num

  def __add__(self,U): # magic method=extend the functionality then we can use the + for objects also
    return self.num+U.num

num1=num(1)
num2=num(2)

res=num1+num2  # this is operator overloading
print(res)
```

```python
class complexnum:
  def __init__(self,x,y):
    self.x=x
    self.y=y

  def __add__(self,C):  #overload
    return self.x+C.x,self.y+C.y

c1=complexnum(1,2)
c2=complexnum(3,4)

res=c1+c2
print(res)
```

```python
class complexnum:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __add__(self, C):  # Overloading +
        return complexnum(self.x + C.x, self.y + C.y)

    def __str__(self):
        return f"({self.x} + {self.y}i)"

c1 = complexnum(1, 2)
c2 = complexnum(3, 4)

res = c1 + c2  # Calls __add__
print(res)  # Calls __str__
```

```python
class democomparisionoverload:
  def __init__(self,x):
    self.x=x

  def __gt__(self,O):
    if(self.x>O.x):
      return True
    else:
      return False

d1=democomparisionoverload(10)
d2=democomparisionoverload(20)

print(d1>d2)
```
