```python
#Histogram
data=[1,2,2,3,3,3,4,4,5]
plt.hist(data,bins=5,color="green")
plt.title("data distribution")
plt.show()
```

In a histogram, **data** appears on the **x-axis**, while the **y-axis** represents the **frequency** (or count) of data points that fall into each bin.

- **x-axis**: Contains the values from the data (in this case, numbers like 1, 2, 3, etc.).
- **y-axis**: Shows how many times each value or range of values appears in the data (the frequency).

To clarify:
- The histogram will divide your data into **bins** (5 in this case), and each bin will represent a range of values. For example, if the bins are set to 5, each bin will contain a specific range of values (like 1-2, 2-3, etc.).
- The **height of each bar** (on the **y-axis**) shows how many data points fall within each bin (i.e., the frequency of data points for that range).

Example:
For your data `[1, 2, 2, 3, 3, 3, 4, 4, 5]` with 5 bins:
- The **x-axis** will have the values divided into 5 bins.
- The **y-axis** will show the frequency (how many numbers fall into each bin).

In simple terms:
- **x-axis**: Values like 1, 2, 3, 4, 5 (the data points).
- **y-axis**: Frequency of how many times those values or bins occur.

labels=labels: Yahan labels list ka reference diya gaya hai, jisme "apple", "banana", aur "orange" hain. Yeh labels chart ke corresponding slices ke saath match karenge.


Jab aap plt.pie(sizes, labels=labels, ...) likhte hain, to labels ko automatically sizes ke corresponding slices ke saath match kiya jata hai. Matlab:

```python
#Pie chart
sizes=[30,20,50]
labels=["apple","banana","orange"]
colors=["gold","lightblue","lightgreen"]
plt.pie(sizes,labels=labels,autopct="%1.1f%%",colors=colors)
plt.title("pie chart")
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data for the plot
x = [1, 2, 3, 4, 5]  # X-axis values
y = [10, 20, 15, 25, 30]  # Y-axis values

# Plotting the data
plt.plot(x, y, marker="o", color="green")  # Green line with circle markers

# Adding title and axis labels
plt.title("Grid Example")  # Title of the plot
plt.xlabel("X axis")  # Label for X-axis
plt.ylabel("Y axis")  # Label for Y-axis

# Enabling the grid
plt.grid(True)  # Show grid lines in the plot

# Saving the plot as a PNG file with 300 DPI resolution
plt.savefig("visual.png", dpi=300)  # Saves the figure with 300 DPI resolution

# Display the plot on the screen
plt.show()  # Show the plot
```

plot type and when to use
* Line plot = for showing trends or continuous data over time

* Bar plot = for comparing different categories

* Histogram = For visualizing data distribution

* Scatter plot = for identifying relationships or correlations between values
* pie chart = for visualizing propertions or percentages

* box plot = for understanding data spread and detecting outliers

Relational plots in seaborn

```python
#Histogram
import seaborn as sns
import matplotlib.pyplot as plt

iris=sns.load_dataset("iris")
sns.histplot(data=iris,x="sepal_length",kde=True,bins=20,color="blue")
plt.title("histogram") #kde = smooth curve,by default frequency count in histogram
plt.show()
```

```python
#Seperate smooth end curve
import seaborn as sns
import matplotlib.pyplot as plt

iris=sns.load_dataset("iris")
sns.kdeplot(data=iris,x="sepal_length",shade=True,hue="species")
plt.title("histogram") #shaded all like blue,orange,green fill
plt.show()
```

```python
#Box plot
#palette = color
import seaborn as sns
import matplotlib.pyplot as plt

iris=sns.load_dataset("iris")
sns.boxplot(data=iris,x="species",y="sepal_length",palette="Set2")
plt.title("box plot")
plt.show()
```

```python
#Violin plot
import seaborn as sns
import matplotlib.pyplot as plt

iris=sns.load_dataset("iris")
sns.violinplot(data=iris,x="species",y="sepal_length",palette="muted",split=True)# split = true no mirror replica
plt.title("violin plot")
plt.show()
# boxplot+belly(density)+peak(outlier strength)
```

Categorical plots

```python
#Bar chart
import seaborn as sns
import matplotlib.pyplot as plt

iris=sns.load_dataset("iris")
sns.barplot(data=iris,x="species",y="sepal_length",ci=None,palette="pastel")
plt.title("avg sepal length by species")
plt.xlabel("species")
plt.ylabel("avg sepal length")
plt.show()
#In your code, ci=None means you are not showing any confidence interval or error bars on the plot.
```

```python
#Count plot
import seaborn as sns
import matplotlib.pyplot as plt

iris=sns.load_dataset("iris")
sns.countplot(data=iris,x="species",palette="Set2")
plt.title("count of species")
plt.xlabel("species")
plt.ylabel("count")
plt.show()
```

```python
#Heat map
import seaborn as sns
import matplotlib.pyplot as plt

iris=sns.load_dataset("iris")
corr=iris.drop("species",axis=1).corr() # .corr() tells strengh of my correlation
sns.heatmap(corr,annot=True,cmap="coolwarm",fmt=".2f") # annot=True, vales inside in it
plt.title("correlation heatmap")
plt.show()
```

```python
#Swarm plot = plots actual number of points
import seaborn as sns
import matplotlib.pyplot as plt

iris=sns.load_dataset("iris")
sns.swarmplot(data=iris,x="species",y="sepal_length",palette="muted")
plt.title("swarm plot")
plt.xlabel
```

```python
#pair plot(possible combinations)
import seaborn as sns
import matplotlib.pyplot as plt
iris=sns.load_dataset("iris")
sns.pairplot(data=iris,hue="species",palette="dark")
plt.title("pair plot")
plt.show()
```

Multiple plotting

```python
#axes (sequence of my graph)
import seaborn as sns
import matplotlib.pyplot as plt

iris=sns.load_dataset("iris")
fig,axes=plt.subplots(1,2,figsize=(5,6))

#plot 1:box plot
sns.boxplot(data=iris,x="species",y="sepal_length",ax=axes[0],palette="pastel")
axes[0].set_title("box plot")

#plot 2
sns.swarmplot(data=iris,x="species",y="sepal_length",ax=axes[1],palette="muted")
axes[1].set_title("swarm plot")

plt.tight_layout()
plt.show()
```

```python
#Regression plot with regplot()
import seaborn as sns
import matplotlib.pyplot as plt

iris=sns.load_dataset("iris")

sns.regplot(data=iris,x="sepal_length",y="sepal_width",line_kws={"color":"red"},scatter_kws={"alpha":0.6}) #trend line alpha close to 0 the point is dark (opeque)
plt.title("regression plot")
plt.show()
#we cant add hue
```

```python
#Regression with Implot()
import seaborn as sns
import matplotlib.pyplot as plt

iris=sns.load_dataset("iris")

sns.lmplot(data=iris,x="sepal_length",y="sepal_width",hue="species",col="species",markers=["o","s","D"])
plt.title("Faceted regression plot") # faceted i grapf splitted
plt.show()
```

```python
#Line plot
import plotly.graph_objects as go

x=[1,2,3,4,5]
y=[2,4,6,8,10]

fig=go.Figure(data=go.Scatter(x=x,y=y,mode="lines+markers",name="sales trend"))
#customise the layout
fig.update_layout(title="line plot",xaxis_title="x-axis",yaxis_title="y-axis",font=dict(size=14))
fig.show()
```

```python
#bar chart
import plotly.graph_objects as go

categories=["A","B","C"]
values=[30,50,20]

fig=go.Figure(data=go.Bar(x=categories,y=values,name="fruit sales"))
fig.update_layout(title="bar chart",xaxis_title="categories",yaxis_title="values",font=dict(size=14))
fig.show()
```

```python
#Scatter plot
import plotly.graph_objects as go

data={"age":[25,30,35,40,45],"salary":[50000,60000,70000,80000,90000],"category":["A","B","C","D","E"]}
fig=go.Figure(go.Scatter(x=data["age"],y=data["salary"],mode="markers",marker=dict(color=['red',"blue","black","green","yellow"]),text=data["category"],name="AGE VS SALARY"))

fig.update_layout(title="scatter plot",xaxis_title="age",yaxis_title="salary",font=dict(size=14))
fig.show()
```

plotly express vs plotly graph

```python
#plotly graph

import plotly.graph_objects as go

#add a bar chart
fig=go.Figure(go.Bar(x=["A","B","C"],y=[30,50,20]))

#customize the layout
fig.update_layout(
    title="Bar Chart Example",
    width=600,  # Set the width of the plot
    height=600 # Set the height of the plot
)
fig.show()
```

```python
#plotly Express
import plotly.express as px

fig=px.bar(x=["A","B","C"],y=[30,50,20],title="bar chart")
fig.show()
```

```python
import plotly.express as px

# Create a bar chart
fig = px.bar(x=["A", "B", "C"], y=[30, 50, 20], title="Bar Chart")

# Customize the size of the plot
fig.update_layout(
    width=400,  # Set the width of the plot
    height=300  # Set the height of the plot
)

# Show the plot
fig.show()
```

```python
#Line chart
import plotly.express as px

fig=px.line(x=[1,2,3,4,5],y=[10,12,13,19,20],title="line chart",labels={"x":"time(days)","y":"sales(units)"})
fig.show()
```
